```python
from PyPDF2 import PdfReader

# Path to the uploaded resume PDF
file_path = "C:/Users/Savio Sunny/OneDrive/Documents/Main project/Main Project Code/jupyter/Resume.pdf"

# Extract text from the PDF
reader = PdfReader(file_path)
extracted_text = ""
for page in reader.pages:
    extracted_text += page.extract_text()

# Output the extracted text
print(extracted_text)
```

```python
import nltk
from nltk.tokenize import word_tokenize
import re

# Function to read multi-word phrases from a text file
def read_multi_word_phrases(file_path):
    with open(file_path, 'r') as file:
        # Read the lines and strip any leading/trailing whitespace
        phrases = file.readlines()
        # Clean up the phrases by removing newlines and extra spaces, converting to lowercase
        phrases = [phrase.strip().lower() for phrase in phrases]
    return phrases

# Path to the text file containing multi-word phrases
file_path = 'multi_word_phrases.txt'

# Read the multi-word phrases from the text file
multi_word_phrases = read_multi_word_phrases(file_path)

# Create a regular expression pattern to match any of the multi-word phrases
multi_word_pattern = re.compile(r'\b(?:' + '|'.join(map(re.escape, multi_word_phrases)) + r')\b')

# Custom tokenizer function to keep multi-word phrases intact
def custom_tokenizer(text):
    # First replace multi-word phrases with a unique placeholder (underscore separator)
    text_with_placeholders = multi_word_pattern.sub(lambda x: x.group(0).replace(" ", "_"), text.lower())
    
    # Now tokenize the text with the placeholders
    tokens = word_tokenize(text_with_placeholders)
    
    # Replace the placeholders back to multi-word phrases by reversing the replacement
    return [token.replace("_", " ") if "_" in token else token for token in tokens]

# Example cleaned text (Ensure you define this text before calling tokenizer)


# Apply the custom tokenizer
tokens = custom_tokenizer(extracted_text)

# Print the tokenized output
print(tokens)
```

STOP WORD REMOVAL

```python
import nltk
from nltk.corpus import stopwords

# Ensure you have the necessary NLTK resources
nltk.download('stopwords')

# Function to remove stopwords from the extracted text
def remove_stopwords(extracted_text):
    """
    Removes common stopwords from the input text using NLTK's stopwords list.
    """
    stop_words = set(stopwords.words('english'))  # Set of English stopwords
    # Split the extracted text into words and filter out the stopwords
    filtered_text = ' '.join([word for word in extracted_text.split() if word.lower() not in stop_words])
    
    return filtered_text



clear_output = remove_stopwords(extracted_text)
print("Text after stopwords removal:\n", clear_output)
```

```python
#REMOVE UNICODE
import re
def remove_unicode_escape(clear_output):
    """
    Removes any unwanted Unicode escape characters from the input text.
    """
    # Remove Unicode escape sequences using a regex pattern
    cleaned_text = re.sub(r'\\[Uu]([0-9A-Fa-f]{4,6})', '', clear_output)
    cleaned_text = re.sub(r'\\[uU]([0-9A-Fa-f]{4})', '', cleaned_text)
   
    
    
    return cleaned_text
cleaned_text=remove_unicode_escape(clear_output)
print(cleaned_text)
```

```python
def remove_extra_whitespaces(text):
    """
    Removes extra whitespaces from the input text (multiple spaces reduced to a single space).
    """
    return re.sub(r'\s+', ' ', text).strip()
cleaned_text=remove_extra_whitespaces(cleaned_text)
```

```python
import string
def remove_punctuation(text):
    """
    Removes punctuation from the input text, except for special characters like '+' in "C++".
    """
    return ''.join([char for char in text if char not in string.punctuation])
cleaned_text=remove_punctuation(cleaned_text)
```
