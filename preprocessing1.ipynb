```python
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer,WordNetLemmatizer

# Ensure you download the stopwords before using this
import nltk
nltk.download('stopwords')

# Initialize stopwords set and stemmer
stop_words = set(stopwords.words('english'))

lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    # Convert text to lowercase
    text = text.lower()
    
    # Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # Tokenize and remove stopwords
    words = text.split()
    filtered_words = [word for word in words if word not in stop_words]
    
  
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_words]
    
    # Join words back into a single string
    processed_text = ' '.join(lemmatized_tokens)
    
    return processed_text

# Example usage
text = extracted_text
processed = preprocess_text(text)

print("Processed text:", processed)
```

```python
import re

# Function to load shortforms from a text file
def load_shortforms(file_path):
    shortform_dict = {}
    with open(file_path, 'r') as file:
        # Read each line from the file and split by ':'
        for line in file:
            if ':' in line:
                shortform, fullform = line.strip().split(':', 1)
                shortform_dict[shortform.strip()] = fullform.strip()
    return shortform_dict

# Function to replace shortforms in the given text with their full forms
def replace_shortforms(text, shortform_dict):
    # Replace each shortform in the text with its full form
    for shortform, fullform in shortform_dict.items():
        # Use regex to replace only whole words (avoiding partial matches)
        text = re.sub(r'\b' + re.escape(shortform) + r'\b', fullform, text)
    return text

# Example usage:
# Load the shortforms from the text file
shortform_dict = load_shortforms("shortform.txt")

# Sample resume text
resume_text =processed

# Replace shortforms in the resume text
updated_resume = replace_shortforms(resume_text, shortform_dict)

# Output the updated resume
print(updated_resume)

```
